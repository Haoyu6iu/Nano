# Surrogate Modelling for Predictive Catalyst Morphology

## 1. Framework and motivations 

The study of the properties of *nanomaterials* is of importance in a wide range of applications (nanotechnologies), including for instance catalysis.

In catalysis, metal nano-clusters, in which there are only a few tens of metal atoms per cluster, enjoy exceptional properties. In this size regime, every single atom is essential (see e.g. Heiz et al., 1999) and extrapolating the properties of nano-clusters from the ones of more massive clusters is generally not possible, as demonstrated for instance for metals like gold (Hutchings et al., 2005). This non-scalability feature implies that the nano regime needs to be analysed independently. 
 
The difference in activity (i.e. catalytic efficiency) between supported nano-clusters depends on their size, their morphology and their interaction with the support. Numerical simulations based on [Density Functional Theory](https://en.wikipedia.org/wiki/Density_functional_theory) (DFT) can be used to explore the properties of nano-clusters, and in particular their [energy landscape](https://en.wikipedia.org/wiki/Energy_landscape). 

Since nano-clusters with low energy (variable `Etot`) are likely to yield efficient catalysts, the search of clusters corresponding to minima (and more specifically, to global minima) of the energy landscapes is of particular interest. Various approaches aiming at directly optimising the DFT energy landscapes have been investigated, based for instance on *particle swarm optimisation* (Avendaño-Franco and Romero, 2016; Call ST et al., 2007), *random search* (Pickard and Needs, 2009 and 2011), *genetic algorithm* (Deaven and Ho, 1995; Davis et al., 2015; Kanters and Donald, 2014; Alexandrova and Boldyrev, 2005; Alexandrova, 2010), *basin hopping* (Zhai et al., 2014; Oganov, 2010; Wales and Doye, 1997), or *simulated annealing* (Wang et al., 2009). However, due to the numerical cost inherent to DFT simulations (the simulation of a single nano-cluster configuration may takes several hours on a modern super-computer), such direct optimisation strategies are extremely timely and expensive from a numerical point of view.  

To accelerate the optimisation process, alternative strategies have been proposed, such as *empirical potentials* (see e.g. Ferrando et al., 2005; Heiles and Johnston, 2013) and *potential energy surface fitting* (see e.g. Ouyang et al., 2015; Zhai et al., 2015; Zhai and Alexandrova, 2016), where the optimisation is not performed directly on the expensive DFT model, but on an easy-to-evaluate approximation of the DFT model. Such approximate models are referred to as [***surogate models***](https://en.wikipedia.org/wiki/Surrogate_model); in the scientific litterature, the terminlogy [computer experiment]( https://en.wikipedia.org/wiki/Computer_experiment) is used to qualify the approximation of expensive numerical simulations.  

In surrogate-model-based approaches, the main challenge is naturally the definition of  accurate approximations of the initial models. This project aims at exploring a data-driven approach, where the machine-learning-based surrogate models are build from the results returned by DFT simulations.

## 2. The data set 

The data set is provided by the School of Chemistry of Cardiff University. It consists of the results of DFT simulations performed with nano-clusters of gold atoms supported by a magnesium-oxyde surface (technically referred to as *MgO(001) surface*); see the figure below.  

The data set consists of two files: `Phys_Descrip.txt` and `Cluster_XYZ.csv`. These two files are intrinsically related: the $i$-th entry of `Phys_Descrip` corresponds to the $i$-th entry of `Cluster_XYZ`, and reciprocally. 

1. An entry (i.e. row) of `Phys_Descrip` contains the values of 16 variables (referred to as "physical descriptors") describing the properties of a nano-cluster. 

2. The corresponding entry of `Cluster_XYZ` contains the coordinates, in $\mathbb{R}^3$, of each of the atoms in the considered nano-cluster (atomic coordinates).  

The data set consists of $n_{\mathrm{obs}}=1{,}839$ nano-clusters of $N$ gold atoms, with $N$ between $1$ and $24$. For a cluster of $N$ atoms, the corresponding row in `Cluster_XYZ` contains $3N$ coordinates: the first $3$ entries are the $(x,y,z)$-coordinates of the first atom in the nano-cluster, the next $3$ entries are the $(x,y,z)$-coordinates of the second atom, etc. 

The nano-clusters are placed at various locations on a same magnesium-oxyde support; the atomic coordinates are all relative to the same coordinate system.

## 3. Objectives

The second objective is to apply some minimisation algorithms to the obtained surrogate models in order to characterise, for a given number of atoms $N$, the properties of the nano-clusters with minimum predicted energies.

## 4. An example of structural fingerprint

Structural fingerprints consist of parametric representations of the atomic coordinates of a nano-cluster that do not depend on the number of atoms in the nano-cluster. In this sense, the provided physiscal descriptors also consist of a fingerprint; they describe both the morphology of the nano-cluster and its interaction with the support.  

The following function `Cluster_FP_X1` gives an example of a naive structural fingerprint; this function takes as entry the atomic coordinates of a nano-cluster, and returns the values of $15$ parameters, independently of the number of atoms $N$ in the nano-cluster. Those $15$ parameter are computed as follows:
* the $3$ coordinates of the barycenter of the nano-cluster;  
* the square roots of the lengths of the three semi-axis of the [ellipsoid] (https://en.wikipedia.org/wiki/Ellipsoid) (i.e. a 3d-[ellipse](https://en.wikipedia.org/wiki/Ellipse)) surrounding the nano-cluster (see the notions of [SVD](https://en.wikipedia.org/wiki/Singular_value_decomposition) and of [PCA](https://en.wikipedia.org/wiki/Principal_component_analysis));  
* the $3\times 3=9$ coordinates of the $3$ main-axes of the surrounding ellipsoid. 

Intuitively, the fingerprint returned by `Cluster_FP_X1` characterises an ellipsoid that approximates the overall shape of the nano-cluster. Such an ellipsoid-based approximation of the morphology is obviously very rought; more sophisticated and precise fingerprints might be defined, and you may thus investigate how to design "efficient" structural fingerprints. 

In `Cluster_FP_X1`, notice that we may actually remove the coordinates of the third main-axis of the ellipsoid since it can be deduced from the other two; the resulting fingerprint would then consist or $12$ variables (instead of $15$). 

```r
Cluster_FP_X1 <- function(Mat_XYZ){
    
    Barycenter <- apply(Mat_XYZ, 2, mean)
    Cent_Coo <- t(t(Mat_XYZ) - Barycenter)    
    
    CoMet = t(Cent_Coo) %*% Cent_Coo  
    Eig = eigen(CoMet) 

    return(c(Barycenter, sqrt(abs(Eig$values)), as.double(Eig$vectors)))
    
}  
```

```r
entry_nb <- 1800
N_atom <- PhyD$N[entry_nb]
Mat_XYZ <- t(matrix(as.double(Cl_XYZ[entry_nb, 1:(3*N_atom)]), 3, N_atom))

noquote('Atomic coordinate:')
print(Mat_XYZ)
noquote('Cluster_FP_X1-fingerprint: ')
Cluster_FP_X1(Mat_XYZ)
```

By applying such fingerprint-generating functions to all the nano-clusters in the data set, we obtain some new features that can next be used to design surrogate models having the ability to deal with various values of $N$.
